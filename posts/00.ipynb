## 1. (결측치) 각 column의 결측값 비율

```python
# 각 column의 결측값 비율 계산
missing_ratio = tidy.isnull().mean()                  # 각 column의 결측값 비율 계산 (결측 개수 / 전체 row 수)
missing_ratio = missing_ratio.sort_values(ascending=False)  # 결측 비율이 높은 순서대로 정렬
missing_ratio = missing_ratio.reset_index()         # Series → DataFrame 변환 (인덱스 초기화)
missing_ratio.columns = ['Column', 'MissingRatio']
print(missing_ratio)
```

## 2. (이상치) 각 column별 이상치 확인

```python
# 수치형 변수만 추출
numeric_cols = tidy.select_dtypes(include=['float64', 'int64']).columns

# 이상치 비율 저장용 리스트
outlier_info = []

# 각 수치형 column에 대해 이상치 비율 계산 (IQR 방식)
for col in numeric_cols:
    Q1 = tidy[col].quantile(0.25)
    Q3 = tidy[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = ((tidy[col] < lower) | (tidy[col] > upper)).sum()
    outlier_ratio = outliers / tidy.shape[0]
    outlier_info.append((col, outliers, outlier_ratio))

# 결과 DataFrame 정리
outlier_df = pd.DataFrame(outlier_info, columns=['Column', 'OutlierCount', 'OutlierRatio'])
outlier_df = outlier_df.sort_values(by='OutlierRatio', ascending=False)

print(outlier_df)
```

```python
# 이상치 비율이 높은 상위 n개 변수만 시각화
import seaborn as sns
import matplotlib.pyplot as plt

# 이상치 비율 높은 상위 n개 변수만 선택
top_n = 5
top_outlier_cols = outlier_df.head(top_n)['Column']

# 박스플롯 시각화
plt.figure(figsize=(12, 6))
for i, col in enumerate(top_outlier_cols, 1):
    plt.subplot(1, top_n, i)
    sns.boxplot(y=tidy[col])
    plt.title(col)
    plt.tight_layout()

plt.show()
```

### column 별로 각 row 비율 확인

```python
def summarize_category(df, column, top_n=None, exclude_na=True):
    """
    범주형 변수 요약 (건수 + 비율)
    
    Parameters:
    - df: DataFrame
    - column: 분석할 열 이름 (문자열)
    - top_n: 상위 n개만 출력 (None이면 전체)
    - exclude_na: NaN 제외 여부
    
    Returns:
    - summary DataFrame
    """
    counts = df[column].value_counts(dropna=exclude_na)
    ratios = df[column].value_counts(normalize=True, dropna=exclude_na).mul(100).round(2)

    summary = pd.DataFrame({
        '건수': counts,
        '비율(%)': ratios
    }).reset_index().rename(columns={'index': column})

    if top_n is not None:
        summary = summary.head(top_n)

    return summary
```

## 4. (기술통계) 각 column의 기술통계 (평균값, 최대값, 최솟값, 최빈값, 분위수 등)

```python
# 수치형 변수만 선택
numeric_cols = tidy.select_dtypes(include=['float64', 'int64']).columns

# 통계 요약 저장용 리스트
stat_list = []

for col in numeric_cols:
    desc = tidy[col].describe()
    mode = tidy[col].mode(dropna=False).values[0]  # 최빈값 (여러 개 중 첫 번째 선택)
    mode_all = tidy[col].mode(dropna=False).tolist()  # 최빈값 리스트 형태로 저장

    stat_list.append({
        'Feature': col,
        'Count': desc['count'],
        'Mean': desc['mean'],
        'Std': desc['std'],
        'Min': desc['min'],
        '25%': desc['25%'],
        'Median (50%)': desc['50%'],
        '75%': desc['75%'],
        'Max': desc['max'],
        'Mode': mode,
        'Mode_all': mode_all
    })

# DataFrame으로 정리
stats_df = pd.DataFrame(stat_list)
stats_df
```

```python
def calc_ratio_by_top_n(df, n):
    top_n = (
        df.dropna(subset=['판매지수', '분류'])
          .sort_values(['분류', '판매지수'], ascending=[True, False])
          .groupby('분류')
          .head(n)
          .copy()
    )
    ratio = (
        top_n.groupby('분류')['추천평유무']
        .mean()
        .round(4)
        .reset_index(name=f'상위{n}추천평비율')
    )
    return ratio

# 추천 비율 계산
ratio30 = calc_ratio_by_top_n(df_rec, 30)
ratio50 = calc_ratio_by_top_n(df_rec, 50)
ratio100 = calc_ratio_by_top_n(df_rec, 100)
ratio200 = calc_ratio_by_top_n(df_rec, 200)
ratio500 = calc_ratio_by_top_n(df_rec, 500)

# 전체 추천 비율 + 전체 책 수
total_stats = (
    df_rec.dropna(subset=['분류'])
         .groupby('분류')
         .agg(
             전체추천평비율=('추천평유무', 'mean'),
             책수=('책ID', 'count')
         )
         .round(4)
         .reset_index()
)

# 병합
merged_all = total_stats.merge(ratio30, on='분류', how='left') \
                        .merge(ratio50, on='분류', how='left') \
                        .merge(ratio100, on='분류', how='left') \
                        .merge(ratio200, on='분류', how='left') \
                        .merge(ratio500, on='분류', how='left') \
                        .sort_values('상위200추천평비율', ascending=False)

# 결과
merged_all
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 복사 및 중복 제거
df = tidy.copy()
df = df.drop_duplicates(subset='책ID').copy()

# 추천 유무 플래그
df['추천유무'] = df['추천평'].notna()

# 출판년도 컬럼 생성
df['출판년도'] = pd.to_datetime(df['출판년월'], errors='coerce').dt.year

# 2020년 이후 필터링 + 결측치 제거
df_plot = df[(df['출판년도'] >= 2020) & df['판매지수'].notna() & df['분류'].notna()].copy()

# 판매지수 중앙값 집계
grouped = df_plot.groupby(['분류', '추천유무'])['판매지수'].median().reset_index()

# 시각화
plt.figure(figsize=(14, 6))
sns.barplot(data=grouped, x='분류', y='판매지수', hue='추천유무')
plt.xticks(rotation=45, ha='right')
plt.title('2020년 이후 분야별 추천유무에 따른 판매지수 중앙값 비교')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 파일 불러오기
df_0517 = pd.read_csv("subctgr_0510_17.csv")
df_0608 = pd.read_csv("subctgr_0608_17.csv")

# 열 이름 통일 (product_count에 날짜명 붙이기)
df_0517 = df_0517.rename(columns={'product_count': '0510_product_count'})
df_0608 = df_0608.rename(columns={'product_count': '0608_product_count'})

# 비교에 필요한 열만 추출
df_0517 = df_0517[['category', 'subcategory', '0510_product_count']]
df_0608 = df_0608[['category', 'subcategory', '0608_product_count']]

# 두 날짜 기준 데이터 병합
merged = pd.merge(df_0517, df_0608, on=['category', 'subcategory'], how='inner')

# 증가량 계산
merged['증가량'] = merged['0608_product_count'] - merged['0510_product_count']

# 증가량 기준 내림차순 정렬
result = merged.sort_values(by='증가량', ascending=False).reset_index(drop=True)

# 결과 확인
result
```
