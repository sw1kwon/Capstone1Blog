# 챕터 2. 정치: 내가 선거 결과를 맞힌 비법

## 1. 요약

이 챕터는 정치 예측이 왜 그토록 자주 빗나가는지와 함께 어떻게 하면 더 나은 예측을 할 수 있는지를 다룬다.

Philip Tetlock의 연구를 통해 직업, 경험, 전공 분야에 상관없이 소위 전문가라고 하는 사람들의 예측이 대부분 초보적인 통계 방법론의 예측보다 못하다는 것을 알 수 있다. Tetlock은 오랜 연구를 통해 전문가들을 하나의 큰 이론으로 세상을 설명하려는 '고슴도치'와 다양한 정보들을 종합해 신중하게 판단하는 '여우'로 분류한다. 때때로 '여우'는 결정을 쉽게 내리지 못해 사람들에게 답답하는 평가를 받기도 하지만 예측은 '고슴도치'보다 훨씬 잘한다.

Nate Silver는 이러한 '여우적 접근'을 바탕으로 자신이 만든 FiveThirtyEight가 어떻게 예측을 하는지, 과거 다른 사람의 잘못된 예측은 왜 틀렸는지, 양적 정보 뿐만 아니라 질적 정보는 어떻게 활용할 수 있는지 등을 설명하며 더 나은 예측을 위해 필요한 점들을 이야기한다.

## 2. 예측과 불확실성 분석

### 저자는 어떤 방법으로 예측하나?

Nate Silver의 FiveThiryEight은 각종 여론조사 결과의 평균을 취하되, 과거에 보인 정확성을 토대로 가중치를 부여하는 단순한 논리로 시작하여 점점 복잡해지고 정교해졌는데 다음 3가지의 '여우'스러운 원칙을 소개한다.

- 확률적으로 생각하라
    - 승패를 단정하지 않고 확률을 사용하여 범위로 제시한다.
- 날마다 새로운 예측을 하라
    - 새로운 정보가 들어오면 과감하게 기존 예측을 수정한다.
- 합의를 구하라
    - 여러 여론조사와 다양한 정보원을 종합하여 집단지성의 이점을 활용한다.

### 과거 정보/경험을 어떻게 활용하는가?

Nate Silver는 과거 조사기관의 정확도를 파악하여 정확했던 기관에는 높은 가중치를, 편향되거나 부정확했던 기관에는 낮은 가중치를 부여한다. 또한 "여론조사는 선거일이 가까워질수록 정해진다"는 역사적 패턴을 바탕으로 선거일까지 남은 기간에 따라 불확실성의 폭을 조절한다.

### 새로운 증거를 어떻게 해석하고 통합하는가?

정확한 예측은 바뀌지 않아야 한다는 잘못된 발상을 깨고 새로운 증거를 통해 과거의 예측이 틀렸다는 것을 알았다면 오늘은 오늘 할 수 있는 최선의 예측을 해야한다. 물론 새로운 예측이 예전의 예측을 무조건 폐기하는 것을 의미하는 것은 아니다.

### 예측의 불확실성을 어떻게 인정하는가?

예측을 확률적으로 표현하는 것이 불확실성을 인정하는 것이다. '고슴도치'들은 불확실성을 예측이 잘못될 경우 들이댈 핑계라고 생각하지만 자기 이론이 완전하지 못하다는 점을 인정해야 실수에서 교훈을 얻을 수 있다. 예측은 훈련을 통해 얼마든지 나아질 수 있다.

### 베이지안 요소

여론조사 결과는 투표자가 적을수록 더 큰 폭으로 춤을 춘다는 내용 (여론 조사 정확도: 하원 선거 < 상원 선거 < 대통령 선거) 을 읽으며 수업 내용 중 사전확률의 분산이 떠올랐다.

| 선거 유형 | 투표자 수 | 여론조사 샘플의 대표성 | 결과 |
|----------|----------|---------------------|------|
| 대통령 선거 | 많음 | 높음 (분산 작음) | 여론조사가 안정적 |
| 상원 선거 | 중간 | 중간 | 중간 정도 변동 |
| 하원 선거 | 적음 | 낮음 (분산 큼) | 여론조사가 춤을 춤 |

- 사전확률의 분산이 작으면 (기존 믿음이 확고하면)
    - 새 데이터가 들어와도 크게 안 흔들림
- 사전확률의 분산이 크면 (기존 믿음이 불확실하면)
    - 새 데이터에 민감하게 반응, 크게 흔들림

정리하면, 투표자 수가 적은 선거일수록 여론조사 데이터의 분산이 크고, 따라 새로운 조사 결과에 예측이 더 민감하게 반응한다. 이는 베이지안 업데이트에서 데이터의 불확실성이 클수록 사후확률의 변동폭이 커지는 원리와 일치한다.

## 3. 비판적 검토

### 저자 주장의 강점과 약점

Tetlock의 오랜 연구에서 나온 '고슴도치'/'여우' 프레임워크와 잘 맞아 떨어지는 Nate Silver의 주장은 그가 이때까지 정확히 예측한 많은 선거들도 고려했을 때 설득력이 있다.

'고슴도치'/'여우'의 간단한 이분법도 좋지만 세상이 워낙 빠르게 변하다보니 '고슴도치'의 빠른 확신이 필요하다거나 이외의 다른 종류의 유형도 나타날 수 있다는 생각이 들었다.

### 반론이나 대안적 해석

주석 37을 보면 질적 자료를 적절하게 활용하는 사례로 나온 Cook Political Report에 대한 소소한 비판으로 예측 대상인 선거들을 지나치게 많이 '반반'으로 분류하고 한 후보가 상당한 수준으로 앞서는 선거구도 '반반'으로 분류한다는 내용이 있다. 이어서 명시적으로 결과를 예측하는 FiveThirtyEight는 Cook Political Repot가 '반반'으로 분류한 50개 중 38개를 적중시켰다고 하는데 Cook Political Report의 이러한 행보를 좋게 볼 여지는 없는지 궁금하여 LLM과 많은 대화를 해봤다.

솔직히 제일 와닿는 내용은 두 기관이 다른 역할을 하고 있으니 사용자가 목적에 따라 선택하면 된다는 내용 뿐이었는데 돌이켜 생각해보니 세상이 결정을 유보하는 '여우'를 답답해하고 뭐라도 결정은 하는 '고슴도치'를 오히려 선호한다는 내용을 Nate Silver와 나에게서 느낄 수 있었다.

### 추가 자료를 통해 발견한 새로운 시각

Tetlock의 후속 연구로, 슈퍼예측가들의 특성을 분석한 'Tetlock, Philip E. & Gardner, Dan. Superforecasting: The Art and Science of Prediction. Crown, 2015.'가 있다는 것을 알았다. 책을 읽지는 못했지만 관련 자료를 찾아보니 슈퍼예측가들이 소유하는 핵심 특징 4가지를 제시하였는데 다음과 같다.

- (1) philosophic outlook
- (2) abilities and thinking style
- (3) method of forecasting
- (4) work ethic

나중에 기회가 된다면 책을 읽으면서 자세한 내용을 확인해야겠다.

### 한계나 누락된 관점

물론 책이 나온지도 시간이 어느정도 흘렀지만, 여론 형성에 크게 영향을 끼치는 SNS, 그와 관련된 알고리즘 문제, AI 등으로 인한 가짜뉴스의 확산 등 빠르게 변하는 세상 속에서 이러한 것들이 어떻게 정치 예측에 영향을 끼칠 수 있는지에 대한 내용이 있었으면 더 재미있었을 것 같다.

## 4. 적용 가능성

### 이 챕터의 통찰을 다른 분야에 적용한다면?

사실 '고슴도치'보다 '여우'처럼 행동하라는 원칙은 어느 분야에도 적용 가능할 듯 싶다. 당장 다음 챕터이자 내가 관심이 있는 스포츠부터 투자, 의료 진단, 기업 전략 등 불확실성이 높은 영역이라면 모두 적용할 수 있다. 예측의 불확실성을 인정하여 확률적으로 생각하면서 새로운 정보를 받아들이며 기존의 정보를 발전시키고 다양한 정보, 사람들을 접한다면 더 나은 의사결정을 할 수 있을 것이다.

### 현실 문제에 어떻게 활용할 수 있을까?

개인이 생각하기에 따라 생각보다 많은 의사결정에 활용할 수 있을 것이다. 당장 내일 우산을 들고 나갈 것인지부터 멀게는 취업을 할지 대학원을 갈지 등 일상 속에서 하나의 확신을 고집하는 것보다 지속적으로 판단을 업데이트하는 과정이 필요하다.
