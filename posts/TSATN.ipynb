# 챕터 2. 정치: 내가 선거 결과를 맞힌 비법

## 1. 요약

이 챕터는 정치 예측이 왜 그토록 자주 빗나가는지와 함께 어떻게 하면 더 나은 예측을 할 수 있는지를 다룬다.

Philip Tetlock의 연구를 통해 직업, 경험, 전공 분야에 상관없이 소위 전문가라고 하는 사람들의 예측이 대부분 초보적인 통계 방법론의 예측보다 못하다는 것을 알 수 있다. Tetlock은 오랜 연구를 통해 전문가들을 하나의 큰 이론으로 세상을 설명하려는 '고슴도치'와 다양한 정보들을 종합해 신중하게 판단하는 '여우'로 분류한다. 때때로 '여우'는 결정을 쉽게 내리지 못해 사람들에게 답답하는 평가를 받기도 하지만 예측은 '고슴도치'보다 훨씬 잘한다.

Nate Silver는 이러한 '여우적 접근'을 바탕으로 자신이 만든 FiveThirtyEight가 어떻게 예측을 하는지, 과거 다른 사람의 잘못된 예측은 왜 틀렸는지, 양적 정보 뿐만 아니라 질적 정보는 어떻게 활용할 수 있는지 등을 설명하며 더 나은 예측을 위해 필요한 점들을 이야기한다.

## 2. 예측과 불확실성 분석

### 저자는 어떤 방법으로 예측하나?

Nate Silver의 FiveThiryEight은 각종 여론조사 결과의 평균을 취하되, 과거에 보인 정확성을 토대로 가중치를 부여하는 단순한 논리로 시작하여 점점 복잡해지고 정교해졌는데 다음 3가지의 '여우'스러운 원칙을 소개한다.

- 확률적으로 생각하라
    - 승패를 단정하지 않고 확률을 사용하여 범위로 제시한다.
- 날마다 새로운 예측을 하라
    - 새로운 정보가 들어오면 과감하게 기존 예측을 수정한다.
- 합의를 구하라
    - 여러 여론조사와 다양한 정보원을 종합하여 집단지성의 이점을 활용한다.

### 과거 정보/경험을 어떻게 활용하는가?

Nate Silver는 과거 조사기관의 정확도를 파악하여 정확했던 기관에는 높은 가중치를, 편향되거나 부정확했던 기관에는 낮은 가중치를 부여한다. 또한 "여론조사는 선거일이 가까워질수록 정해진다"는 역사적 패턴을 바탕으로 선거일까지 남은 기간에 따라 불확실성의 폭을 조절한다.

### 새로운 증거를 어떻게 해석하고 통합하는가?

정확한 예측은 바뀌지 않아야 한다는 잘못된 발상을 깨고 새로운 증거를 통해 과거의 예측이 틀렸다는 것을 알았다면 오늘은 오늘 할 수 있는 최선의 예측을 해야한다. 물론 새로운 예측이 예전의 예측을 무조건 폐기하는 것을 의미하는 것은 아니다.

### 예측의 불확실성을 어떻게 인정하는가?

예측을 확률적으로 표현하는 것이 불확실성을 인정하는 것이다. '고슴도치'들은 불확실성을 예측이 잘못될 경우 들이댈 핑계라고 생각하지만 자기 이론이 완전하지 못하다는 점을 인정해야 실수에서 교훈을 얻을 수 있다. 예측은 훈련을 통해 얼마든지 나아질 수 있다.

### 베이지안 요소

여론조사 결과는 투표자가 적을수록 더 큰 폭으로 춤을 춘다는 내용 (여론 조사 정확도: 하원 선거 < 상원 선거 < 대통령 선거) 을 읽으며 수업 내용 중 사전확률의 분산이 떠올랐다.

| 선거 유형 | 투표자 수 | 여론조사 샘플의 대표성 | 결과 |
|----------|----------|---------------------|------|
| 대통령 선거 | 많음 | 높음 (분산 작음) | 여론조사가 안정적 |
| 상원 선거 | 중간 | 중간 | 중간 정도 변동 |
| 하원 선거 | 적음 | 낮음 (분산 큼) | 여론조사가 춤을 춤 |

- 사전확률의 분산이 작으면 (기존 믿음이 확고하면)
    - 새 데이터가 들어와도 크게 안 흔들림
- 사전확률의 분산이 크면 (기존 믿음이 불확실하면)
    - 새 데이터에 민감하게 반응, 크게 흔들림

정리하면, 투표자 수가 적은 선거일수록 여론조사 데이터의 분산이 크고, 따라 새로운 조사 결과에 예측이 더 민감하게 반응한다. 이는 베이지안 업데이트에서 데이터의 불확실성이 클수록 사후확률의 변동폭이 커지는 원리와 일치한다.

## 3. 비판적 검토

### 저자 주장의 강점과 약점

Tetlock의 오랜 연구에서 나온 '고슴도치'/'여우' 프레임워크와 잘 맞아 떨어지는 Nate Silver의 주장은 그가 이때까지 정확히 예측한 많은 선거들도 고려했을 때 설득력이 있다.

'고슴도치'/'여우'의 간단한 이분법도 좋지만 세상이 워낙 빠르게 변하다보니 '고슴도치'의 빠른 확신이 필요하다거나 이외의 다른 종류의 유형도 나타날 수 있다는 생각이 들었다.

### 반론이나 대안적 해석

주석 37을 보면 질적 자료를 적절하게 활용하는 사례로 나온 Cook Political Report에 대한 소소한 비판으로 예측 대상인 선거들을 지나치게 많이 '반반'으로 분류하고 한 후보가 상당한 수준으로 앞서는 선거구도 '반반'으로 분류한다는 내용이 있다. 이어서 명시적으로 결과를 예측하는 FiveThirtyEight는 Cook Political Repot가 '반반'으로 분류한 50개 중 38개를 적중시켰다고 하는데 Cook Political Report의 이러한 행보를 좋게 볼 여지는 없는지 궁금하여 LLM과 많은 대화를 해봤다.

솔직히 제일 와닿는 내용은 두 기관이 다른 역할을 하고 있으니 사용자가 목적에 따라 선택하면 된다는 내용 뿐이었는데 돌이켜 생각해보니 세상이 결정을 유보하는 '여우'를 답답해하고 뭐라도 결정은 하는 '고슴도치'를 오히려 선호한다는 내용을 Nate Silver와 나에게서 느낄 수 있었다.

### 추가 자료를 통해 발견한 새로운 시각

Tetlock의 후속 연구로, 슈퍼예측가들의 특성을 분석한 'Tetlock, Philip E. & Gardner, Dan. Superforecasting: The Art and Science of Prediction. Crown, 2015.'가 있다는 것을 알았다. 책을 읽지는 못했지만 관련 자료를 찾아보니 슈퍼예측가들이 소유하는 핵심 특징 4가지를 제시하였는데 다음과 같다.

- (1) philosophic outlook
- (2) abilities and thinking style
- (3) method of forecasting
- (4) work ethic

나중에 기회가 된다면 책을 읽으면서 자세한 내용을 확인해야겠다.

### 한계나 누락된 관점

물론 책이 나온지도 시간이 어느정도 흘렀지만, 여론 형성에 크게 영향을 끼치는 SNS, 그와 관련된 알고리즘 문제, AI 등으로 인한 가짜뉴스의 확산 등 빠르게 변하는 세상 속에서 이러한 것들이 어떻게 정치 예측에 영향을 끼칠 수 있는지에 대한 내용이 있었으면 더 재미있었을 것 같다.

## 4. 적용 가능성

### 이 챕터의 통찰을 다른 분야에 적용한다면?

사실 '고슴도치'보다 '여우'처럼 행동하라는 원칙은 어느 분야에도 적용 가능할 듯 싶다. 당장 다음 챕터이자 내가 관심이 있는 스포츠부터 투자, 의료 진단, 기업 전략 등 불확실성이 높은 영역이라면 모두 적용할 수 있다. 예측의 불확실성을 인정하여 확률적으로 생각하면서 새로운 정보를 받아들이며 기존의 정보를 발전시키고 다양한 정보, 사람들을 접한다면 더 나은 의사결정을 할 수 있을 것이다.

### 현실 문제에 어떻게 활용할 수 있을까?

개인이 생각하기에 따라 생각보다 많은 의사결정에 활용할 수 있을 것이다. 당장 내일 우산을 들고 나갈 것인지부터 멀게는 취업을 할지 대학원을 갈지 등 일상 속에서 하나의 확신을 고집하는 것보다 지속적으로 판단을 업데이트하는 과정이 필요하다.

# 챕터 3. 야구: 야구 경기는 왜 모든 '예측'의 모델이 되는가?

## 1. 요약

이 챕터는 야구에서 어떻게 데이터 기반으로 예측을 할 수 있는지 그리고 통계와 인간의 판단이 어떻게 공존할 수 있는지에 대해 다룬다.

Nate Silver는 자신이 개발한 야구 예측 모델인 PECOTA(Pitcher Empirical Comparison and Optimization Test Algorithm) 를 중심으로 이야기를 전개한다. 타자와 투수, 팀의 성적 예측 뿐만 아니라 유망주의 미래 성적까지 예측을 하며 통계적 접근과 스카우터의  판단이 어떤 강점과 한계를 갖는지 설명한다.

핵심은 양적 정보(통계적 접근)과 질적 정보(스카우터의 판단)을 모두 활용해야 한다는 것이다. PECOTA와 같은 예측 모델은 선수의 정신적 능력 등을 파악할 수 없고, 스카우터의 직관만으로는 많은 데이터에서 신호를 찾아내기 어렵다.

## 2. 예측과 불확실성 분석

### 저자는 어떤 방법으로 예측하나?

Nate Silver는 좋은 야구 예측 시스템으로 다음 세 가지 사항을 제시한다.

- 각 선수의 통계자료가 갖는 맥락의 의미를 설명하기
    - 홈구장 효과, 리그수준 등
- 실력과 운을 분리한기
    - 투수가 아무리 잘 던져도 타자가 득점을 못하면 이길 수 없음
- 각 선수가 나이를 먹어감에 따라서 성적이 어떻게 진화하는지 이해하기
    - 노화곡선(aging curve) 이해하기

### 과거 정보/경험을 어떻게 활용하는가?

주석 18을 보면 Nate Silver는 프로야구가 현대적 경기로 발전한 2차 세계대전 발발 이후의 모든 메이저리그 시즌 기록과 두 선수를 비교할 수 있는 알고리즘을 탑재한 데이터 베이스를 개발하였다.

PECOTA의 핵심은 Bill James가 개발한 '유사도 점수'의 확장이다. 과거 회고에만 머물렀던 James와 달리 Silver는 '최근접 이웃 분석(nearest neighbor analysis)' 기법을 사용하여 키, 몸무게 같은 스카우터 영역의 변수까지 고려하여 이를 미래 예측에 적용하였다.

### 새로운 증거를 어떻게 해석하고 통합하는가?

PECOTA의 선수 예측 시스템의 핵심 요소 중 하나로 기준선 예측(Baseline Forecasts)에서 새로운 증거를 어떻게 해석하고 통합하는지 확인할 수 있다. 새로운 증거라고 볼 수 있는 최근 성적을 더 크게 반영하는 가중 평균을 만들고 ‘Regression to the mean’을 활용하여 진짜 실력을 추정한다.

참고로 PECOTA에서는 이외에도 마이너리그 성적을 메이저리그 성적으로 환산하는 작업 (Major-League Equivalencies)과 예측 대상 선수와 체격, 포지션, 성적 패턴 등에 대한유사도 점수를 활용한 커리어 경로 조정(Career-Path Adjustment)을 함께 진행하여 선수의 기량을 예측한다.

### 예측의 불확실성을 어떻게 인정하는가?

PECOTA는 통계 내용이 비슷한 선수들 각자가 맞이할 결과를 다양한 스펙트럼의 범위('최선의 사례', '최악의 사례', '가장 닮은꼴 사례')로 제시하여 불확실성이 일어날 수 있음을 인정하였다.

또 뚜렷하게 구분되는 26개의 노화곡선이 각기 다른 선수에게 1개씩 적용된다는 가설을 세운 Gary Huckabay의 시스템을 활용하면서도 Nate Silver는 모든 선수를 하나의 패턴에 억지로 끼워맞추지 않았다.

### 베이지안 요소

| 구분                       | Beta–Binomial (이항 + 베타)                                                               | Dirichlet–Multinomial (다항 + 디리클레)                                                                          |
| ------------------------ | ------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------------------------- |
| 다루는 데이터 형태               | "성공/실패" 2범주                                                                           | "여러 범주" $K$개                                                                                               |
| 한 번의 관측(시행)              | $X\in\{0,1\}$                                                                           | $Z\in\{1,\dots,K\}$                                                                                          |
| $n$번 모은 데이터              | $k$ = 성공 횟수                                                                           | $\mathbf{x}=(x_1,\dots,x_K)$, $\sum x_i=n$                                                                 |
| 미지의 파라미터                 | $\theta$ (성공확률, 0~1)                                                                  | $\mathbf{p}=(p_1,\dots,p_K)$ (범주확률벡터, 합=1)                                                                 |
| 사전분포(prior)              | $\theta\sim \mathrm{Beta}(\alpha,\beta)$                                              | $\mathbf{p}\sim \mathrm{Dirichlet}(\alpha_1,\dots,\alpha_K)$                                               |
| 가능도(likelihood)          | $k\mid\theta\sim \mathrm{Binomial}(n,\theta)$                                         | $\mathbf{x}\mid\mathbf{p}\sim \mathrm{Multinomial}(n,\mathbf{p})$                                          |
| 사후분포(posterior)          | $\theta\mid k\sim \mathrm{Beta}(\alpha+k,\ \beta+n-k)$                                | $\mathbf{p}\mid \mathbf{x}\sim \mathrm{Dirichlet}(\alpha_1+x_1,\dots,\alpha_K+x_K)$                        |
| 업데이트 규칙(한 줄)             | 성공이면 $\alpha{+}{=}1$, 실패면 $\beta{+}{=}1$                                            | 범주 $i$가 나오면 $\alpha_i{+}{=}1$                                                                             |
| 사후 평균(포인트 추정)            | $\mathbb{E}[\theta\mid k]=\dfrac{\alpha+k}{\alpha+\beta+n}$                           | $\mathbb{E}[p_i\mid \mathbf{x}]=\dfrac{\alpha_i+x_i}{\sum_j\alpha_j+n}$                                    |
| "가상 표본(pseudo-count)" 해석 | $\alpha$=가상 성공, $\beta$=가상 실패, $\alpha+\beta$=사전 강도                                   | $\alpha_i$=가상 범주 카운트, $\sum\alpha_i$=사전 강도                                                                 |
| 사후예측분포(미래 데이터)           | 다음 $m$번 성공수 $\tilde{k}\sim \mathrm{Beta\text{-}Binomial}(m,\alpha+k,\beta+n-k)$ | 다음 $m$번 카운트 $\tilde{\mathbf{x}}\sim \mathrm{Dirichlet\text{-}Multinomial}(m,\boldsymbol\alpha+\mathbf{x})$ |
| 언제 쓰면 딱 맞나               | 한 가지 사건 확률만 추정할 때                                                                     | 여러 결과가 동시에 경쟁할 때                                                                                           |
| 야구 예시                    | "삼진 여부", "볼넷 여부", "스트라이크 여부" 같은 0/1                                                   | "타석 결과 분해(삼진/볼넷/인플레이아웃/1B/2B/3B/HR …)"                                                                     |

LLM과 베이지안 관련으로 떠오르는 내용에 대해 대화하다가 정리한 내용이다. 사실 수식적으로 다 안다고 할 수는 없지만 conjugate prior여서 계산이 용이하다는 것과 야구 예시에서 왜 적용되는지는 확실히 알 수 있다. 또 관련 논문들도 있어서 공부하다가 기회가 되면 읽어볼 생각이다.

## 3. 비판적 검토

### 저자 주장의 강점과 약점

1만 건의 선수-시즌 데이터베이스를 구축하여 만든 PECOTA가 여러 예측에 성공한 점과 양적 접근의 한계를 인정한 점이 설득력 있게 다가온다.

하지만 요즘 맹활약을 펼치고 있는 투타 겸업의 선수 Shohei Ohtani처럼 비교할 만한 과거 사례가 없는 경우 유사도 점수의 근본적 한계가 존재하고 아무리 뛰어난 선수라도 부상으로 인해 출전을 못 한다면 의미가 없는데 부상에 대한 내용이 제시되지 않았다.

### 반론이나 대안적 해석

위에서 부상에 대한 내용이 제시되지 않은 점을 약점으로 뽑았지만 이에 대한 반론도 떠오르고 해서 LLM과 대화 후 몇 가지 내용을 정리해보았다.

- 역할 분담의 관점
    - 부상예측은 구단의 피지컬 팀의 전문 영역이므로 부상 위험 평가는 의료진의 질적 판단에 맡기고
    - PECOTA는 "건강하다면 어떤 성적을 낼 것인가"에 집중
- 부상은 본질적으롤 예측 불가능에 가까움
    - 부상의 상당 부분은 우연적 사건(충돌,사고)이나 복잡한 생체역학적 요인이 얽혀 있어 데이터로 예측하기 어려움
    - "예측할 수 없는 것을 억지로 예측하지 않겠다"는 여우적 신중함의 표현
- 간접적으로 반영되어 있음
    - 부상이 잦았던 선수 유형(특정 체격, 포지션, 플레이 스타일)은 이미 비교군의 커리어 경로에 반영되어 있음
- 현재 보완되고 있음
    - MLB는 Statcast, Biomechanics 분석 등을 통해 위험 예측하려는 시도를 하고 있음 

책에서 강조하고 나도 동의하는 양적 정보와 질적 정보의 결합의 측면에서 볼 때 충분히 타당하다고 생각되는 내용들이다.

### 추가 자료를 통해 발견한 새로운 시각

선수를 26개의 노화곡선 중 하나에 무조건 넣는 Huckabay의 노화곡선이 신기해서 주석 14의 관련 기사를 찾아보았다. 솔직히 자신의 노화곡선에 대해 옹호하는 내용이 많을 줄 알았는데 그런 내용보다 SPSS, RMSE, 자유도 등 익숙한 단어도 많아서 재미있게 읽었고 본인이 인정하는 한계들도 정리해보았다.

- 비전형적 선수 문제
    - 비교 선수가 없는 경우
- 데이터 범위의 한계
    - 부상 정보, 코칭 정보, 상대한 투수 정보
- 자유도 문제
    - 데이터를 900개 범주로 나누려면, 각 범주에 충분한 데이터가 있어야 모델 신뢰 가능
- 출장 시간 예측의 어려움
    - 출장 시간을 예측하려 하면, 시스템이 별로인 선수에게 0이나 마이너스 타석을 주려 함

나도 올해 여름부터 해외축구 5대 리그의 매주 경기 후 stat들을 꾸준히 크롤링하고 있는데 이 데이터를 어떻게 쓸지는 아직 모르겠지만 위의 내용들이 나중에 도움이 될 것 같다.

### 한계나 누락된 관점

Biomechanics에 대한 내용도 있으면 좋았을 것 같긴하지만 책이 나온지도 시간이 흘렀고 데이터로만 접근하기도 어려운 분야라 어쩔 수 없다고 생각한다.

## 4. 적용 가능성

### 이 챕터의 통찰을 다른 분야에 적용한다면?

PECOTA의 '유사도 점수' 개념은 다양한 분야에 적용 가능하다. 축구, 농구와 같은 다른 종목의 스포츠에도 적용할 수 있고 의료 분야에서는 환자의 나이, 증상, 검사 결과 등이 유사환 과거 환자들의 치료 결과를 분석하여 예측하는 연구가 진행중이다.

### 현실 문제에 어떻게 활용할 수 있을까?

챕터를 읽으며 제일 와 닿은 부분은 양적 정보와 질적 정보의 결합이다. Dustin Pedroia가 초반에 성적이 좋지 않았을 때도 꾸준히 자기의 모습을 유지하여 좋은 성적을 거둔 점, 스카우터 John Sanders가 '정신적 도구상자'(준비성과 노동 윤리, 집중과 초점, 경쟁심과 자신감, 스트레스 관리와 겸손)을 중요하게 보고 다니는 점 등 좁게는 스포츠, 넓게는 현실에서 사람을 평가하는데 있어서 KPI와 같은 양적 정보가 절대적이지는 않은 것 같다. 그래서 사람을 상대하든 내가 어떤 행동을 하든 양적 정보만 신경쓰지 말고 질적 정보도 신경쓰면서 사는 자세가 필요하다고 생각한다.

# 최종 성찰

## 이 과제를 하며 내 독서/사유 능력이 어떻게 변했는가?

일단 평소에 독서를 잘 하지 않다가 이렇게 책을 읽게 되니 혼자 생각해보고 고민해보는 시간을 가질 수 있어서 유익하였다. 물론 발표와 보고서를 해야하기에 평소보다 더 생각하거나 정보를 찾으면서 읽느라 시간도 오래 걸렸지만 이러한 과정이 낭비라고 생각하지는 않는다. 특히 한 챕터씩만 읽었는데도 불구하고 LLM을 활용하니 다양한 정보들을 마음만 먹으면 끊임 없이 접할 수 있었다. 흔히 책 한권만 읽은 사람이 무섭다고 하는데 그것은 '고슴도치' 같이 읽었을 때 이야기인 것 같다. LLM을 활용하며 '여우'의 자세로 비판적 사고를 가지고 읽는다면 책 한권으로도 편협한 사고에 빠지지 않을 수 있다고 생각한다. 

## 가장 어려웠던 점과 극복 방법

핑계지만 책을 읽는 시간을 내는 것이 제일 어려웠다. 평소에 습관이 되있지도 않고 책을 집중해서 오래 읽을 시간을 만들려다보니 오히려 읽지 않게 되었다. 그래서 오래 읽지 못하더라도 조금씩 자주 읽으려고 하였고 한 번 읽기 시작하니 오래 읽게 되는 경우도 있었다.

## AI를 어떻게 활용했고, 어떤 도움이 되었나?

일단 책을 읽으면서 떠오르는 생각을 그냥 질문해도 대답해줘서 그것 자체로 생각을 정리할 수 있어서 도움이 되었다. 또 베이지안 통계학 수업인만큼 읽는 내용을 베이지안 쪽으로 생각해보려고 하였는데 잘 모르겠는 내용도 입력을 하고 대화를 하면서 알 수 있었다. 보고서에 있는 베이지안 관련 내용은 머리 속으로 떠오른 생각을 그냥 질문한 다음 대화를 통해 정리한 내용들이다. 챕터 2에서는 다음과 같이 대화를 시작하였다

여론조사 결과는 투표자가 적을수록 더 큰 폭으로 춤을 춤
정확도: 하원 선거 여론조사 < 상원 선거 여론조사 < 대통령 선거 여론조사

이거를 베이지안 요소로 간단하게 볼 수 없을까?
베이지안에서 업데이트할 때 분산의 크기에 따라 막 달라지는 거 있지 않아? 알기 쉽게 설명해줘

이 외에도 비판적으로 책을 읽기 위해 질문을 하거나 일부러 내 생각과 다른 입장에 대해 물어보는 등 다양한 사고를 하는데 도움을 얻을 수 있었다. 또 책을 읽으며 궁금한 추가 자료를 찾고 주석 등에서 찾은 자료를 정리하는데도 많은 도움을 받았다.

## 앞으로 복잡한 텍스트를 읽을 때 달라질 점

복잡한 텍스트를 읽으면 한 번에 이해해야겠다는 생각에 오히려 잘 읽지 못하는 경우가 많은데 LLM을 활용한다면 일단 처음에 이해되는 수준으로 한 번 읽은 다음 그 상태에서 질문을 하면서 조금씩 알아갈 수 있을 것 같다. 당연한 것이 생각나지 않더라도 쉽게 물어보고 답변을 받을 수 있기 때문에 복잡한 텍스트에 대한 부담감이 줄어들 것 같다.
